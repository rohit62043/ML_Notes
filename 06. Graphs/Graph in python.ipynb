```python
#We will use matplotLib for plotting graph
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5];
y=[2,4,6,8,10];

#plt.scatter(x,y)
plt.plot(x,y,"r^") #Third parameter is String which first character decide color and 2 onward line type--> -(contineous),--(discontious),--.,p(pentagon),o(dot)
plt.grid()
plt.show()
```

```python
#plot function x^3
import numpy as np;
x=np.arange(1,10,0.1);
y=x**3;

# plt.plot(x,y,"r--");
plt.plot(x,y,color="pink",marker="^",linewidth=0.2) #marker will be along with continous line
plt.ylabel("x^3");
plt.xlabel("x");
plt.title("polynomial graph");
plt.show();
```

```python
#plot function x^3 and x^2
import numpy as np;
x=np.arange(1,10,0.1);
y=x**3;
y1=x**2;

# plt.plot(x,y,"r--");
plt.plot(x,y,"r-",linewidth=4,label="x^3") #marker will be along with continous line
plt.plot(x,y1,"b-",linewidth=4,label="x^2")
plt.ylabel("y");
plt.xlabel("x");
plt.title("polynomial graph");
plt.legend();
#we can also chage axis size
plt.axis([0,30,0,1000]); #[xStart,xEnd,yStart,yEnd]
#we can also add comment at particular point
plt.text(10,500,"Hello Wordl",fontsize=12);
plt.show();
```

```python
a=[3,4,5]
plt.plot(a,linewidth=2.5) #it will treate a as y and for x it will take a range from 0-n
plt.grid()
plt.show()
```

```python
#pie Graph
labels=["A","B","C","D"]
sizes=[3,4,6,2]
colors=["blue","yellow","red","green"];
#highlighting class
explode=[0.1,0,0,0];
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct="%.2f%%",counterclock=False,startangle=120) #we can also pass a function inside autopct
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```
